```python
import pandas as pd
```

```python
df = pd.read_csv("movies_metadata.csv", encoding='ISO-8859-1', on_bad_lines='skip')
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['budget'], df['revenue'], alpha=0.5, color='blue')
plt.title('Budget vs Revenue', fontsize=14)
plt.xlabel('Budget (in dollars)', fontsize=12)
plt.ylabel('Revenue (in dollars)', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['vote_average'], bins=10, color='green', edgecolor='black')
plt.title('Distribution of Movie Ratings', fontsize=14)
plt.xlabel('Vote Average', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
df['year'] = df['release_date'].dt.year

plt.figure(figsize=(10,6))
df.groupby('year').size().plot(kind='line', marker='o', color='orange')
plt.title('Number of Movies Released Over Time', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Movies Released', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['runtime'].dropna(), bins=20, color='red', edgecolor='black')
plt.title('Distribution of Movie Runtime', fontsize=14)
plt.xlabel('Runtime (in minutes)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['vote_average'], df['revenue'], alpha=0.5, color='blue')
plt.title('Revenue vs Vote Average', fontsize=14)
plt.xlabel('Vote Average', fontsize=12)
plt.ylabel('Revenue (in dollars)', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10,6))
correlation_matrix = df[['budget', 'revenue', 'runtime', 'vote_average', 'vote_count']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Movie Features', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
top_movies = df.nlargest(10, 'revenue')

plt.figure(figsize=(10,6))
plt.barh(top_movies['original_title'], top_movies['revenue'], color='teal')
plt.title('Top 10 Movies by Revenue', fontsize=14)
plt.xlabel('Revenue (in dollars)', fontsize=12)
plt.ylabel('Movie Title', fontsize=12)
plt.tight_layout()
plt.show()
```
